````python
 #Removing the values from the column 'State' and making new dataframe, 'statewise_df'

statewise_df = df.drop(df[(df.State == 'Total (Uts)') | (df.State == 'Total (All India)') | 
               (df.State == 'Total (States)')].index)
statewise_df = statewise_df.drop(statewise_df[statewise_df['Total'] == 0].index)

````

````python
#Renaming the 'Type' which has almost same meaning
statewise_df.replace('Bankruptcy or Sudden change in Economic', 
           'Bankruptcy or Sudden change in Economic Status', inplace=True)
statewise_df.replace('By Other means (please specify)', 'By Other means', inplace=True)
statewise_df.replace('Not having Children(Barrenness/Impotency',
           'Not having Children (Barrenness/Impotency)', inplace=True)
````

````python
statewise_df = statewise_df.set_index(pd.Index([i for i in range(0,101110)]))
statewise_df
````

````python
import seaborn as sns  #https://seaborn.pydata.org/generated/seaborn.set_style.html
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
#We use rc i.e a runtime configuration (rc) containing the default styles for every plot element you create
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
#Yearwise Death

#Creating a series of total suicides according to year

yearwise_total_suicide = statewise_df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)

#plotting this in a line graph
yearwise_total_suicide.plot(kind="line", figsize = (9,5), color = 'r')
plt.title('Yearwise Total Suicides')
plt.xlabel('Year')
plt.ylabel('Number of suicides');
````

Observation  : Yearwise Suicides cases are increases from 2001 to 2010.In 2011 it increases at large number 677159 but after that suicide rate seems to be decreasing . 

````python
# Statewise suicide case 

#Creating a series of total suicides on the basis of states
statewise_total_suicide = statewise_df.groupby("State")["Total"].sum()
print('Total number of suicides in each State : ', statewise_total_suicide)

#plotting the above data
filter_state = pd.DataFrame(statewise_total_suicide.reset_index()) 
sort_state = filter_state.sort_values('Total',ascending = True)
f, ax = plt.subplots(figsize=(15, 4))
#sns.barplot(y = 'Total', x = 'State',data = sort_state,edgecolor = '0.5')
sns.barplot(y = 'Total', x = 'State',data = sort_state)
plt.xticks(rotation=75);
plt.show()
````

**Statewise Suicide Cases**:
    - Maharashtra has the highest number of suicides with 901,945 cases.
    - Other states with high suicide rates include West Bengal, Andhra Pradesh, and Tamil Nadu.
    - Lakshadweep has the lowest number of suicides with only 50 cases.

Q3 Suicides According To The Causes In Each Year

````python
#Creating a series of total suicides on the basis of causes
causewise_total_suicide = statewise_df.groupby("Type")["Total"].sum()
print('Total number of suicides by cause: ', causewise_total_suicide)

#plotting the above data
filter_year = pd.DataFrame(causewise_total_suicide.reset_index()) 
sort_year = filter_year.sort_values('Total', ascending=True)
f, ax = plt.subplots(figsize=(15, 4))
sns.barplot(y='Total', x='Type', data=sort_year)
plt.xticks(rotation=75);
plt.show()
````

Suicide According To Causes in Each Year

````python
#Filtering the data according to 'type_code','Total' and 'Year'
filter_type_code = pd.DataFrame(statewise_df.groupby(["Type_code","Year"])["Total"].sum()).reset_index()

#Plotting the above data
sns.catplot(x="Type_code", y="Total",hue="Year", kind="bar", data=filter_type_code,height=7.5, aspect=10/7.5);
plt.title('Suicide According To Causes In Each Year');
plt.show()
````

````python
# Grouping the data by Year and Type and summing the Total column
causes_yearwise = statewise_df.groupby(['Year', 'Type'])['Total'].sum().unstack()

# Plotting the data
causes_yearwise.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Suicides According to the Causes in Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend(title='Cause', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````

**Suicides According to Causes in Each Year**:
    - The most common causes of suicides include family problems, illness, and love affairs.
    - The number of suicides due to family problems has been consistently high across the years.

Q. Suicides By The Educational Background

````python
# Creating a pivot table for suicides based on education and gender
education_gender_suicide = statewise_df.pivot_table(values='Total', index='Gender', columns='Type_code', aggfunc='sum')

# Plotting the data
education_gender_suicide[['Education_Status']].plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Suicides Based on Education and Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Suicides')
plt.legend(title='Education Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````

````python
# Plotting a pie chart for suicides based on education and gender
education_gender_suicide['Education_Status'].plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8))
plt.title('Suicides Based on Education and Gender')
plt.ylabel('')
plt.show()
````

````python
# Filter the dataframe for educational reasons
education_reasons = statewise_df[statewise_df['Type_code'] == 'Education_Status']

# Group by the 'Type' column and sum the 'Total' column
education_reasons_total = education_reasons.groupby('Type')['Total'].sum()

# Print the total number of suicides for each educational reason
print(education_reasons_total)

# Plotting the data
education_reasons_total.plot(kind='bar', figsize=(12, 6))
plt.title('Suicides by Educational Reasons')
plt.xlabel('Educational Reason')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.show()
````

````python
# Group by the 'Type' and 'Gender' columns and sum the 'Total' column
education_reasons_gender_total = education_reasons.groupby(['Type', 'Gender'])['Total'].sum().unstack()

# Print the total number of suicides for each educational reason by gender
print(education_reasons_gender_total)

# Plotting the data
education_reasons_gender_total.plot(kind='bar', figsize=(12, 6))
plt.title('Suicides by Educational Reasons and Gender')
plt.xlabel('Educational Reason')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.show()
````

````python
#filtering the data as per the educational background
education_status = pd.DataFrame(statewise_df[statewise_df["Type_code"]=="Education_Status"].groupby(["Type","Gender"])["Total"].sum()).reset_index()
education_status

#plotting the above data
g = sns.catplot(y="Type", x="Total",hue="Gender", kind="bar", data=education_status,height=7, aspect=11.7/7);
sns.set_style('whitegrid');
plt.show()
````

**Suicides Based on Education and Gender**:
    - Individuals with no education and those with primary education have the highest suicide rates.
    - Both males and females with no education or primary education are more prone to committing suicide.

Q.Suicides By Age Groups

````python
# Grouping the data by Age_group and Gender and summing the Total column
age_gender_suicide = statewise_df.groupby(['Age_group', 'Gender'])['Total'].sum().unstack()

# Plotting the data
age_gender_suicide.plot(kind='bar', figsize=(12, 6))
plt.title('Suicides by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
````

````python
#Age Distributiion
statewise_Age_df = statewise_df.drop(statewise_df[statewise_df.Age_group == '0-100+'].index)

counts = statewise_Age_df['Age_group'].value_counts().sort_index()
print(counts)

counts.plot(kind='bar', title='Age Distribution',figsize=(7,5))
plt.xticks(rotation=0)
plt.show()
````

````python
#Age Distributiion
# Categorize by gender
statewise_Age_df = statewise_df.drop(statewise_df[statewise_df.Age_group == '0-100+'].index)
statewise_Age_df['Gender'] = statewise_Age_df['Gender'].astype('category')

counts = statewise_Age_df['Age_group'].value_counts().sort_index()
print(counts)

counts.plot(kind='bar', title='Age Distribution',figsize=(7,5))
plt.xticks(rotation=0)
plt.show()
````

**Suicides by Age Group and Gender**:
    - The age group 15-29 has the highest number of suicides, followed by the age group 30-44.
    - Both males and females in the age group 15-29 are more prone to committing suicide.
**Age Distribution**:
    - The age distribution shows that the majority of suicides occur in the age group 15-29.
    - The number of suicides decreases with increasing age.
